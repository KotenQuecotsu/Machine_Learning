Типизация значений

```python
def convert_to_float(series):
    if(type(series) != float):
        converted_series = series.replace(',','').replace('%','').replace('$','')
        return float(converted_series)
    else:
        if not np.isnan(series):
            return float(series)
        else:
            return float(np.nan_to_num(series))
def convert_to_int(series):
    if(type(series) != int and type(series) != float):
        converted_series = series.replace(',','').replace('%','').replace('$','')
        return int(converted_series)
    else:
        if not np.isnan(series):
            return int(series)
        else:
            return int(np.nan_to_num(series))
```

```python
for series in df.columns:
    if df[series].dtype == 'object':
        try:
            df[series] = df[series].apply(convert_to_int)
        except:
            try:
                df[series] = df[series].apply(convert_to_float)
            except:
                df[series] = df[series]
df['Official language'] = df['Official language'].astype('category')
df.dtypes
```

Группировка


```python
df.groupby('Official language').mean(numeric_only=True)
```

```python
agregation_dict ={
}
df_for_agregation = df[['Official language','Population']]
for item in df_for_agregation['Official language']:
    agregation_dict[item] = {'values':[],'agregated': 0,'median': 0, 'mean': 0}
for item in df_for_agregation['Official language'].unique():
    for index in df.index[df['Official language'] == item].tolist():
        agregation_dict[item]['values'].append(df['Population'][index])

for language in agregation_dict:
    agregation_dict[language]['values'].sort()
    values = agregation_dict[language]['values']
    values_length = len(values)
    agregation_dict[language]['agregated'] = sum(values)
    if values_length != 0:
        agregation_dict[language]['mean'] = sum(values)/values_length
        if values_length % 2 == 0:
            agregation_dict[language]['median'] = (values[values_length//2-1] + values[values_length//2])/2
        else:
            agregation_dict[language]['median'] = values[values_length//2-1]

agregation_dict
```
